```python
#Is the world happy?
plt.title('Is the world Happy?')
sns.distplot(df['Ladder score'], hist=True, kde=True, 
             bins=40, color = 'orange', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 1})
#Most countries have a happiness rating b/w 4 and 7
```

```python
happiness=[]
for row in df['Ladder score']:
    if row > 2.0 and row <= 4.296:
        happiness.append(0)
    if row > 4.296 and row <= 6.069:
        happiness.append(1)
    if row > 6.069:
        happiness.append(2)
df['Happiness'] = happiness
```

```python
x = (df['Happiness'].value_counts())
l1 = []
l1.append(x[0])
l1.append(x[1])
l1.append(x[2])
my_colors = ['orange','gold','yellow']
label = ["Unhappy", "Somewhat Happy", "Happy"]
plt.title('Is the world Happy?')
plt.pie(l1, labels = label, colors=my_colors, explode=[0.2, 0, 0], autopct='%1.1f%%')
plt.show()
#the pie chart reinforces the findings from the histogram
```
